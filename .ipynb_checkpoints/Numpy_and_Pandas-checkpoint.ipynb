```python
import numpy as np
import pandas as pd
```

# Creating a DataFrame from an Array

```python
data = np.array([[1, 2], [3, 4], [5, 6]])
print(data)
```

```python
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2'])
print(df)
```
